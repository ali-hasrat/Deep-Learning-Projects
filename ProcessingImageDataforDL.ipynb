````python
# importing the image module from matplotlib library
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
````

````python
# loading an image module through matplotlib.image module
img = mpimg.imread('/content/dog_image.jpg')
````

````python
type(img)
````

````python
# shape of image in form of array
img.shape
````

````python
# print the numpy array the image contain
print(img)
````

````python
# displaying the image from numpy array

img_plot = plt.imshow(img)
plt.show()
````

Resizing the Image using Pillow Library

````python
from PIL import Image
````

````python
img = Image.open('/content/dog_image.jpg')
img_resized = img.resize((200,200))
````

````python
img_resized.save('dog_image_resized.jpg')
````

````python
# displaying the image from numpy array
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
````

````python
#now checking the shape of new image
img_res.shape
````

**Converting RGB image to Grayscale image using OpenCV**

````python
# importing the OpenCV library
import cv2
````

````python
img = cv2.imread('/content/dog_image.jpg')
````

````python
type(img)
````

````python
img.shape
````

````python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
````

````python
type(grayscale_image)
````

````python
grayscale_image.shape
````

Thus, we saw this image has color channel is only one not three like RGB image
- It's been converted just due to make process easier as it only contain one channel.
